```python
import cv2 as cv
import scipy.io as sio

def read_mat(filename):
    return sio.loadmat(filename)
```

## 1

```python
kitchen_noise = read_mat('data/kitchen_noise.mat')['kitchen_noise']
kitchen_noise.shape
```

```python
kitchen_noise.min(), kitchen_noise.max()
```

```python
kitchen_noise = cv.normalize(kitchen_noise, None, 0, 255, cv.NORM_MINMAX)
kitchen_noise = kitchen_noise.astype('uint8')

kitchen_noise.min(), kitchen_noise.max()
```

```python
import matplotlib.pyplot as plt

plt.imshow(kitchen_noise)
plt.axis('off')
plt.show()
```

```python
r, g, b = cv.split(kitchen_noise)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(r, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(g, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(b, cmap='gray')
plt.axis('off')

plt.show()
```

```python
r_changed = r[750:810,365:500] 


plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.imshow(r_changed, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.bar(r_changed.flatten(), range(len(r_changed.flatten())))

plt.show()
```

```python
filtered_r = cv.medianBlur(r, 9)

plt.figure(figsize=(15, 5))

plt.imshow(filtered_r, cmap='gray')
plt.axis('off')
plt.show()
```

```python
g_changed = g[750:810,365:500] 

plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.imshow(g_changed, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.bar(g_changed.flatten(), range(len(g_changed.flatten())))

plt.show()
```

```python
filtered_g = cv.medianBlur(g, 5)

plt.figure(figsize=(15, 5))

plt.imshow(filtered_g, cmap='gray')
plt.axis('off')
plt.show()
```

```python
b_changed = b[750:810,365:500] 


plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.imshow(b_changed, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.bar(b_changed.flatten(), range(len(b_changed.flatten())))

plt.show()
```

```python
filtered_b = cv.medianBlur(b, 9)

plt.figure(figsize=(15, 5))

plt.imshow(filtered_b, cmap='gray')
plt.axis('off')
plt.show()
```

```python
merged_filtered = cv.merge([filtered_b, filtered_g, filtered_r])

cv.imwrite('output/kitchen_noise_filtered.jpg', merged_filtered)
```
