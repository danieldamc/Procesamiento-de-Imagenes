```python
import cv2 as cv
import scipy.io as sio

def read_mat(filename):
    return sio.loadmat(filename)
```

## 1

```python
kitchen_noise = read_mat('data/kitchen_noise.mat')['kitchen_noise']
kitchen_noise.shape
```

```python
kitchen_noise.min(), kitchen_noise.max()
```

```python
kitchen_noise = cv.normalize(kitchen_noise, None, 0, 255, cv.NORM_MINMAX)
kitchen_noise = kitchen_noise.astype('uint8')

kitchen_noise.min(), kitchen_noise.max()
```

```python
import matplotlib.pyplot as plt

plt.imshow(kitchen_noise)
plt.axis('off')
plt.show()
```
